```python
import wikipedia
import csv
import random
import time
import os  

def obtener_articulos_wikipedia(cantidad, idioma="eu", archivo_salida=r"C:\Users\IÑIGO\Desktop\Universidad\PLN\DATASETEUSKERA\articulosTrain.csv"):
  
    wikipedia.set_lang(idioma)

    directorio = os.path.dirname(archivo_salida)
    if not os.path.exists(directorio):
        os.makedirs(directorio) 

    articulos = []  

    for _ in range(cantidad):
        try:
            titulo = wikipedia.random()
            
            contenido = wikipedia.page(titulo).content
            articulos.append({"titulo": titulo, "contenido": contenido})
            print(f"Artículo añadido: {titulo}")
            time.sleep(random.uniform(1, 3))
        except Exception as e:
            print(f"Error al descargar el artículo: {e}")

    with open(archivo_salida, mode="w", encoding="utf-8", newline="") as archivo:
        escritor = csv.DictWriter(archivo, fieldnames=["titulo", "contenido"])
        escritor.writeheader()
        escritor.writerows(articulos)

    print(f"Se han guardado {len(articulos)} artículos en el archivo {archivo_salida}.")

#1500 articulos de wikipedia
obtener_articulos_wikipedia(3500)
```

```python
import pandas as pd
import wikipedia

def get_ores_score(title, lang='eu'):
    """
    Obtiene la puntuación ORES para un artículo de Wikipedia dado su título.

    Args:
        title (str): Título del artículo.
        lang (str): Código del idioma de la Wikipedia (por defecto, euskera 'eu').

    Returns:
        float: Puntuación ORES (predicción de calidad), o None si no se encuentra.
    """
    import requests

    base_url = f"https://ores.wikimedia.org/v3/scores/{lang}wiki/"

    try:
        # Obtener el ID de revisión más reciente del artículo usando la librería wikipedia
        wikipedia.set_lang(lang)
        page = wikipedia.page(title)
        revid = page.revision_id

        if not revid:
            return None

        # Obtener la predicción de calidad de ORES
        response = requests.get(f"{base_url}?models=articlequality&revids={revid}").json()
        scores = response.get(lang + 'wiki', {}).get('scores', {}).get(str(revid), {}).get('articlequality', {})
        if 'score' in scores and 'prediction' in scores['score']:
            return scores['score']['prediction']
        else:
            return None

    except Exception as e:
        print(f"Error obteniendo la puntuación ORES para {title}: {e}")
        return None

def add_ores_scores_to_csv(input_csv, output_csv, lang='eu'):
    """
    Lee un CSV con títulos de artículos de Wikipedia y contenido, 
    añade una columna con las puntuaciones ORES basadas en los títulos.

    Args:
        input_csv (str): Ruta al archivo CSV de entrada.
        output_csv (str): Ruta al archivo CSV de salida con puntuaciones ORES.
        lang (str): Código del idioma de la Wikipedia (por defecto, euskera 'eu').

    """
    try:
        # Cargar el CSV
        df = pd.read_csv(input_csv)

        if 'titulo' not in df.columns:
            raise ValueError("El archivo CSV debe tener una columna 'Título'.")

        # Crear una columna para las puntuaciones
        scores = []
        for index, title in enumerate(df['titulo']):
            print(f"Procesando {index + 1}/{len(df)}: {title}")
            scores.append(get_ores_score(title, lang=lang))
        df['ORES_Score'] = scores

        # Guardar el resultado en un nuevo archivo CSV
        df.to_csv(output_csv, index=False)
        print(f"Archivo con puntuaciones ORES guardado en: {output_csv}")

    except Exception as e:
        print(f"Error procesando el archivo CSV: {e}")

# Ruta al archivo CSV
input_csv = "C:\\Users\\IÑIGO\\Desktop\\Universidad\\PLN\\DATASETEUSKERA\\articulosTrain.csv"
output_csv = "C:\\Users\\IÑIGO\\Desktop\\Universidad\\PLN\\DATASETEUSKERA\\articulosTrain_with_ORES.csv"

# Ejecutar la función
add_ores_scores_to_csv(input_csv, output_csv)


```
